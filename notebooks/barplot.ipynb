```python
import matplotlib.pyplot as plt
import numpy as np

# MLP
weights1 = [0.166579260111175, 0.1667361510446617, 0.1666995399654974, 0.16635029710561625, 0.166654399079931, 0.16698035269311864]
weights2 = [0.08917960513705195, 0.31474027218708067, 0.05026835345984282, 0.04006133793367836, 0.11222925052712286, 0.3935211807552233]

# Données d'exemple
categories = ['Downstairs', 'Jogging', 'Sitting', 'Standing', 'Upstairs', 'Walking']

# Largeur des barres
largeur_barre = 0.35

# Création du bar plot pour la première série de données
plt.bar(np.arange(len(categories)), weights1, width=largeur_barre, label='Use of weighted sampler')

# Création du bar plot pour la deuxième série de données (décalé sur la droite)
plt.bar(np.arange(len(categories)) + largeur_barre, weights2, width=largeur_barre, label='Use of random sampler')

# Ajout de labels et de titre
plt.xlabel('Labels')
plt.ylabel('Percentage of observation')
#plt.title('Superposition de Bar Plots')
plt.xticks(np.arange(len(categories)) + largeur_barre / 2, categories)

# Ajout de la légende
plt.legend()

plt.savefig('MLP_weights.png')
# Affichage du graphique
plt.show()
```
